#Collabrative Filtering#

```python
df=ratings.groupby('User-ID').count()['Book-Rating']>200
```

```python
indexs=df[df].index
```

```python
filter_rating=ratings[ratings['User-ID'].isin(indexs)]
filter_rating
```

```python
index=filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
z=index[index].index
```

```python
title_rating=filter_rating[filter_rating['Book-Title'].isin(z)]
title_rating
```

```python
Final_df=title_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
Final_df.head()
```

```python
Final_df=Final_df.fillna(0)
Final_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
score=cosine_similarity(Final_df)
score
```

```python
score.shape
```

```python
def recommend(book):
  data=[]
  index=np.where(Final_df.index==book)[0][0]
  boo=sorted(list(enumerate(score[index])),key=lambda x:x[1],reverse=True)[1:11]
  for i in boo:
    l1=[]
    temp=books[books['Book-Title']==Final_df.index[i[0]]]
    l1.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
    l1.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
    l1.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))

    data.append(l1)
  return data
```

```python
recommend('The Da Vinci Code')
```
